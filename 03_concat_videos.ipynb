```python
import subprocess
import tempfile
from pathlib import Path


def concat_videos(video_paths, output_path):
    """
    使用 ffmpeg concat demuxer 拼接视频
    要求：所有视频编码参数一致（分辨率、codec、fps等）
    """

    # 转成 Path 对象
    video_paths = [Path(p).resolve() for p in video_paths]
    output_path = Path(output_path).resolve()

    # 创建临时文件保存视频列表
    with tempfile.NamedTemporaryFile(mode="w", delete=False, suffix=".txt") as f:
        list_file = Path(f.name)

        for path in video_paths:
            # ffmpeg 要求每行格式：file 'path'
            f.write(f"file '{path.as_posix()}'\n")

    try:
        cmd = [
            "ffmpeg",
            "-y",                         # 覆盖输出文件
            "-f", "concat",
            "-safe", "0",
            "-i", str(list_file),
            "-c", "copy",                  # 无损拼接
            str(output_path)
        ]

        print("Running:", " ".join(cmd))
        subprocess.run(cmd, check=True)

        print(f"✅ 拼接完成: {output_path}")

    finally:
        list_file.unlink(missing_ok=True)


# =========================
# 示例
# =========================

video_paths = [
    "/home/tianqi/D/01_Projects/06_ffmpeg/fiftyone/Getting Started with FiftyOne Computer Vision Workshop_ Section 1 - Introduction and Housekeeping.mp4",
    "/home/tianqi/D/01_Projects/06_ffmpeg/fiftyone/Getting Started with FiftyOne Computer Vision Workshop_ Section 2 - Overview.mp4",
    "/home/tianqi/D/01_Projects/06_ffmpeg/fiftyone/Getting Started with FiftyOne Computer Vision Workshop_ Section 3 - Installation & Environments.mp4",
    "/home/tianqi/D/01_Projects/06_ffmpeg/fiftyone/Getting Started with FiftyOne Computer Vision Workshop_ Section 4 - Basic Concepts.mp4",
    "/home/tianqi/D/01_Projects/06_ffmpeg/fiftyone/Getting Started with FiftyOne Computer Vision Workshop_ Section 5 - Hands On Exploration.mp4",
    "/home/tianqi/D/01_Projects/06_ffmpeg/fiftyone/Getting Started with FiftyOne Computer Vision Workshop_ Section 6 - Next Steps.mp4",
]

concat_videos(video_paths, "Getting Started with FiftyOne Computer Vision Workshop.mp4")
```
